```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def show_image(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

cap = cv2.VideoCapture('/Users/faridsoroush/Desktop/IMG_8484.mp4') 

# Read the first frame
ret, prev_frame = cap.read()
prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
warp_mode = cv2.MOTION_TRANSLATION
warp_matrix = np.eye(2, 3, dtype=np.float32)
number_of_iterations = 5000
termination_eps = 1e-10
criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, number_of_iterations, termination_eps)

count = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    (cc, warp_matrix) = cv2.findTransformECC(prev_gray, frame_gray, warp_matrix, warp_mode, criteria)
    stabilized_frame = cv2.warpAffine(frame, warp_matrix, (frame.shape[1], frame.shape[0]), flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
    prev_gray = frame_gray.copy()

    # Show every 100th frame
    if count % 100 == 0:
        print(f"Frame: {count}")
        show_image(frame)
        show_image(stabilized_frame)

    count += 1

cap.release()
```
